```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(df.drop('fire_occurred', axis=1), df.fire_occurred)
```

```python
rf.predict([[32,51,2,4.3]])
```

```python
import pickle
with open('rf3.pkl', 'wb') as f:
    pickle.dump(rf, f)
```
